```python
import pandas as pd
import numpy as np
import os
```

```python
all = pd.read_csv("all.csv")
all.head()
```

```python
exempt = pd.read_csv("exempt.csv")
exempt.head()
```

```python
holiday = pd.read_csv("holiday.csv")
holiday.head()
```

```python
type(holiday["Postcode"])
```

```python
exempt_set = set(exempt["Postcode"])
holiday_set = set(holiday["Postcode"])
```

```python
all["exempt_tax"] = np.where(all["postcode"].isin(exempt_set), "exempt", "tax")
all["holiday"] = np.where(all["postcode"].isin(holiday_set), True, False)
```

```python
all.head(n=20)
```

```python
all.to_csv("all_exempt_holiday.csv", index=False)
```

```python
BASE_DIR = os.getcwd()
vic_locality_path = "ASGS_Coding_Indexes_2016\\2016_LOCALITY_2016_SA2_CODING_INDEX.csv"

locality_names = ["LOCALITY_NAME", "POSTCODE"]
vic_locality = pd.read_csv(os.path.join(BASE_DIR, vic_locality_path), usecols=locality_names)
vic_locality.head()
```

```python
# vic_locality.POSTCODE.sort_values(inplace=True)
vic_locality["INC"] = vic_locality.groupby("POSTCODE").cumcount()
vic_locality.head(n=20)
```

```python
vic_locality = vic_locality.pivot(index = ["POSTCODE"], values = "LOCALITY_NAME", columns = "INC")
```

```python
vic_locality.head()
```

```python
vic_locality['LOC']=vic_locality.apply(lambda row: row.dropna().tolist(), axis=1)
```

```python
vic_locality.to_csv("vic_locality_pivot.csv")
```
